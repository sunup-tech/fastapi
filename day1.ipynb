```python
from openai import OpenAI
import base64

# OpenAI client 생성
client = OpenAI()

# 이미지 파일 base64 인코딩
def encode_image(path):
    with open(path, "rb") as img_file:
        return base64.b64encode(img_file.read()).decode("utf-8")

image_base64 = encode_image("capibara.jpg")

# 멀티모달 요청 (텍스트 + 이미지)
response = client.responses.create(
    model="gpt-4o-mini",
    input=[
        {
            "role": "user",
            "content": [
                {"type": "input_text",
                "text": "이 이미지가 무엇인지 설명해 주세요."},
                {
                    "type": "input_image",
                    "image_url": f"data:image/png;base64,{image_base64}"
                }
            ]
        }
    ],
    max_output_tokens=300
)

print(response.output_text)
```

```python
from openai import OpenAI
import base64

client = OpenAI()

# Step 1에서 생성한 이미지 불러오기
with open("step1.png", "rb") as image_file:

    response = client.images.edit(
        model="gpt-image-1",
        image=image_file,
        prompt="배경을 공동묘지로 바꿔줘."
    )

# Base64 이미지 추출
image_base64 = response.data[0].b64_json

# PNG 파일로 저장
with open("step2.png", "wb") as f:
    f.write(base64.b64decode(image_base64))

print("step2.png 저장 완료")
```

```python
# pip install openai
import base64
import json
from openai import OpenAI

client = OpenAI(
    api_key=api_key,
    base_url="https://api.upstage.ai/v1/document-classification"
)

def encode_img_to_base64(img_path):
    with open(img_path, 'rb') as img_file:
        img_bytes = img_file.read()
        base64_data = base64.b64encode(img_bytes).decode('utf-8')
        return base64_data

filepath = "abcd.png"
base64_data = encode_img_to_base64(filepath)

response = client.chat.completions.create(
    model="document-classify",
    messages=[
        {
            "role": "user",
            "content": [
                {
                    "type": "image_url",
                    "image_url": {"url": f"data:application/octet-stream;base64,{base64_data}"}
                }
            ]
        }
    ],
    response_format={
        "type": "json_schema",
        "json_schema": {
            "name": "document-classify",
            "schema": {
                "type": "string",
                "oneOf": [
                    {"const": "invoice", "description": "상세 품목별 요금과 청구 정보가 포함된 상업 송장"},
                    {"const": "receipt", "description": "구매 거래 내역이 표시된 영수증"},
                    {"const": "contract", "description": "법률 계약서 또는 계약 문서"},
                    {"const": "cv", "description": "이력서 또는 커리큘럼 비타"},
                    {"const": "others", "description": "기타"}
                ]
            }
        }
    }
)
print(response)

```
