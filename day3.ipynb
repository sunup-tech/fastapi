```python
from dotenv import load_dotenv
import os

load_dotenv(override=True)
os.getenv('OPENAI_API_KEY')
default_model = os.getenv('OPENAI_DEFAULT_MODEL')
gocamping_api_key = os.getenv('GOCAMPING_API_KEY')
gemini_default_model = os.getenv('GEMINI_DEFAULT_MODEL')
os.getenv('GEMINI_API_KEY')

from openai import OpenAI

client = OpenAI()
print(gemini_default_model)
```

```python
from pykrx import stock
from datetime import datetime, timedelta
import pandas as pd


def get_stock_price(ticker,day):
    end_date = datetime.now()
    start_date = end_date - timedelta(days=day)

    df = stock.get_market_ohlcv_by_date(start_date.strftime('%Y%m%d'),
    end_date.strftime('%Y%m%d'), ticker)
    
    return df

ticker = '005930'
df = get_stock_price(ticker, 30)
```

```python
df.describe().T
```

```python
latest_price = df['종가'].iloc[-1]
highest_price = df['고가'].max()
lowest_price = df['저가'].min()
average_price = df['종가'].mean()

analysis = (
    f"최근 30일 동안 삼성전자의 종가 데이터를 분석한 결과는 다음과 같습니다\n"
    f"Latest Price: {latest_price}\n"
    f"Highest Price in the period: {highest_price}\n"
    f"Lowest Price in the period: {lowest_price}\n"
    f"Average Price in the period: {average_price}"
    f"이 데이터를 기반으로 간단한 분석과 전망을 제공해주세요"
)   

ret = client.responses.create(
    model=default_model,
    instructions= '당신은 주식분석 전문가입니다. 다음 데이터를 기반으로 분석과 전망을 제공해주세요',
    input=analysis
)
print(ret.output_text)
```

## ChatPromptTemplate

```python
from langchain_core.prompts import ChatPromptTemplate 

template = ChatPromptTemplate.from_messages([
    ('system', '당신은 친절한 챗봇입니다. 답변은 간단히 5줄 이내로 해주세요'),
    ('user','{input}')])
prompt = template.format_messages(input='프로야구 3개팀에 대해 알려줘')
ret = llm.invoke(prompt)
print(ret.content)
```

```python
chain = template | llm
ret = chain.invoke({'input':'프로야구 3개팀에 대해 알려줘'})
```

```python
시스템 메세지 : 훌륭한 가이드, 고객 최적의 맞춤 여행 일정 수립을 도와줍니다
user 메세지: {when} {place} {travel}을 계획하고 있습니다
ai 메세지 : 내가 바로 {place}의 여행 전문가입니다. 
user 메세지 : {questions} #해당 지역의 여행지 5군데를 추천해주세요 
```

```python
from langchain_core.prompts import ChatPromptTemplate 

template = ChatPromptTemplate.from_messages([
    ('system', '당신은 훌륭한 가이드입니다. 고객 최적의 맞춤 여행 일정 수립을 도와줍니다'),
    ('user', '{when}에 {place}로 {travel}을 계획하고 있습니다'),
    ('ai', '내가 바로 {place}의 여행 전문가입니다.'),
    ('user', {questions})
])

prompt = template.format_messages(when='여름', place='제주도', travel = '휴가', questions= '해당 지역의 여행지 5군데를 추천해주세요')
print(prompt)
ret = llm.invoke(prompt)
print(ret.content)
```
