```python
from langchain_core.chat_history import InMemoryChatMessageHistory
memory = InMemoryChatMessageHistory()
memory.add_message(('system', '당신은 여행 전문가로 고객의 여행 일정에 도움을 줍니다'))


question = ['부산 여행에서 딱 한 곳만 가봐야 한다면 어떤 곳인지 알려주세요',
    '부산역에서 그곳에 가는 교통편을 알려주세요',
    '그 근처에서 먹을 만한 음식점들을 추천해주세요',
    '그 근처 카페도 소개해주세요'
]

for q in question:
    memory.add_user_message(q)
    ret = llm.invoke(memory.messages)
    memory.add_ai_message(ret.content)
    print(q)

for msg in memory.messages:
    print(msg)
```

### SummaryMemory
- 긴 대화 요약 : 그 앞의 데이터들은 요약 + 가장 최근 것은 그대로 

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.prompts import MessagesPlaceholder
from langchain_core.runnables.history import RunnableWithMessageHistory

cfg1 = {'configurable': {'session_id': 'user1'}}
cfg2 = {'configurable': {'session_id': 'user2'}}
cfg3 = {'configurable': {'session_id': 'user3'}}

store = {}

def get_memory(session_id):
    if session_id not in store:
        store[session_id] = InMemoryChatMessageHistory()
    return store[session_id]

def get_summary_memory(session_id):
    h = get_memory(session_id)
    if len(h.messages) > 5:
        summary = llm.invoke(f'다음 데이터를 요약해줘 \n {h.messages[:-4]}').content
        h.messages = [summary] + h.messages[-4:]
    return h

prompt = ChatPromptTemplate.from_messages([
    ('system', '당신은 친절한 한국어 챗봇이야'),
      MessagesPlaceholder('history'), 
      ('user', '{input}')
])

chain = RunnableWithMessageHistory(
    prompt | llm,
    get_summary_memory,
    input_messages_key='input',
    history_messages_key='history')
```

```python
chain.invoke({'input':'안녕하세요 챗봇님'},config=cfg2)
chain.invoke({'input':'나는 홍길동입니다'},config=cfg2)
chain.invoke({'input':'서울을 관광하고 있습니다. 서울 음식점 한 곳만 알려주세요'},config=cfg2)
chain.invoke({'input':'그 곳의 제일 인기 있는 메뉴는 무엇인가요?'},config=cfg2)
chain.invoke({'input':'서울역에서 그 곳까지 대중교통으로 가려면 어떻게 가야하나요?'},config=cfg2)
chain.invoke({'input':'내 이름 기억하니?'},config=cfg2)
```

```python
from datetime import datetime

from langchain_core.runnables import RunnableLambda
from langchain_core.prompts import ChatPromptTemplate
from langchain_openai import ChatOpenAI
from langchain_core.output_parsers import StrOutputParser

# 오늘 날짜를 반환하는 Runnable 생성
date_runnable = RunnableLambda(
    lambda _: {
        "today": datetime.now().strftime("%Y-%m-%d"),
        "question": "오늘 점심 뭐 먹을까?"
    }
)

# 프롬프트 템플릿
prompt = ChatPromptTemplate.from_messages([
    ("system", "당신은 친절한 음식 추천 어시스턴트입니다."),
    ("human", "오늘은 {today}. {question}")
])

# LLM 설정
llm = ChatOpenAI(
    model="gpt-4o-mini",
    temperature=0.7
)

# 문자열 파서
parser = StrOutputParser()

# 체인 구성 (LCEL 방식)
chain = date_runnable | prompt | llm | parser

# 실행
response = chain.invoke({})
print(response)
```

```python
from langchain_core.tools import tool
from langchain_openai import ChatOpenAI
from langchain_core.messages import HumanMessage, ToolMessage

@tool
def add(x: int, y:int) -> int:
    '''두 숫자를 더하는 함수입니다.'''
    return x + y

@tool
def strlen(s: str) -> int:
    '''  문자열의 길이를 반환하는 함수입니다'''
    return len(s)

tools = [add, strlen]
tools_dict = {tool.name : tool for tool in tools} #툴(함수) 이름(문자열)과 툴(함수)실제함수: {'add' : add}
tools_dict

llm = ChatOpenAI(model=default_model)
llm_with_tools = llm.bind_tools(tools) #add 툴을 사용할 수 잇는 llm 객체 

#query = '5와 7을 더해줘'
query = "'무궁화꽃이 피었습니다'의 문자열 길이를 알려줘"
#1차 llm 호출 - tool 판단           #llm호출-> tool 호출 여부 판단 여부 결정 ->tool_calls(0), tool calls(x)
ret1 = llm_with_tools.invoke(query)
print(ret1.tool_calls)

if ret1.tool_calls:                #tool_call이 존재하면 해당 tool을 실행시킴
    #tool을 호출 결과 result로 반환
    call = ret1.tool_calls[0]
    result = tools_dict[call['name']].invoke(call['args']) # result = add(5, 7)

    #2차 호출
    messages = [('user',query),    #user의 질문, ret1(tool_calls), ToolMessage(call_id, content)
                ret1,
                ToolMessage(tool_call_id=call['id'], content=result)]
    ret2 = llm_with_tools.invoke(messages)
    print(ret2.content)

else:
    print(ret1.content)




```

```python
# tool 한 개 정의

from dotenv import load_dotenv

from langchain_openai import ChatOpenAI

from langchain_core.tools import tool

from langchain_core.messages import HumanMessage
from langchain_tavily import TavilySearch

load_dotenv()

# Tool 정의

#@tool은 일반 Python 함수를 Agent가 사용할 수 있는 Tool로 등록해 주는 데코레이터.

#Agent는 함수명, 타입 힌트, Docstring을 읽어서 언제 이 Tool을 사용할지 판단한다.

@tool

def add(a: int, b: int) -> int:

    """두 숫자를 더합니다."""

    return a + b

@tool

def get_word_length(word: str) -> int:

    """단어나 문장의 정확한 글자 수(길이)를 계산할 때 사용합니다."""

    return len(word)

search = TavilySearch(max_results=3)
@tool
def web_search(query:str) -> str:
    """웹 검색을 수행하는 도구입니다"""
    return search.run(query)
# LLM 생성

llm = ChatOpenAI(model=default_model)  #gpt-4o

tools=[add, get_word_length, web_search]

tools_dict = { t.name: t for t in tools }

# Tool 연결

llm_with_tool = llm.bind_tools(tools)

# 질문

# query = '5와 7을 더하기'

# query = "'만나서 반갑습니다'는 글자 수가 몇 개인지 알려줘"

query = "서울과 도쿄의 날씨를 비교해줘" # 오류 남

 

messages = [('user', query)]

for _ in range(10):

    res = llm_with_tool.invoke(messages)

    messages.append(res) 

    if not res.tool_calls:

        print(res.content)

        break

    

    for call in res.tool_calls:

        result = tools_dict[call['name']].invoke(call['args'])

        print(f"\nTool '{call['name']}' 실행 결과: {result}")

        messages.append(ToolMessage(tool_call_id=call['id'], content=str(result)))    

        print(messages)

print(res.content)
```

```python
# tool 한 개 정의

from dotenv import load_dotenv

from langchain_openai import ChatOpenAI

from langchain_core.tools import tool

from langchain_core.messages import HumanMessage
from langchain_tavily import TavilySearch
from langgraph.prebuilt import create_react_agent


load_dotenv()

# Tool 정의

#@tool은 일반 Python 함수를 Agent가 사용할 수 있는 Tool로 등록해 주는 데코레이터.

#Agent는 함수명, 타입 힌트, Docstring을 읽어서 언제 이 Tool을 사용할지 판단한다.

@tool

def add(a: int, b: int) -> int:

    """두 숫자를 더합니다."""

    return a + b

@tool

def get_word_length(word: str) -> int:

    """단어나 문장의 정확한 글자 수(길이)를 계산할 때 사용합니다."""

    return len(word)

search = TavilySearch(max_results=3)
@tool
def web_search(query:str) -> str:
    """웹 검색을 수행하는 도구입니다"""
    return search.run(query)

# LLM 생성
llm = ChatOpenAI(model=default_model)  #gpt-4o
tools=[add, get_word_length, web_search]



# 질문
query = "오늘 한국 대 멕시코 축구 경기 결과를 예상해줘 몇 대 몇 이런 식으로" # 오류 남

agent = create_react_agent(llm, tools=tools)
result = agent.invoke({'messages':query})
```
